```python
from pathlib import Path
```

```python
input_stm = "CV_reference.stm"
output_stm = "CV_reference_update.stm"
dict_tsv = "speaker_mapping.tsv"
```

```python
#this function can be used to map common voice's long speaker labels to 'spkr{x}', with a tsv dictionary for the old and new speaker labels

def create_speaker_mapping_and_replace(input_stm_path, output_stm_path, dict_tsv_path):
    speaker_map = {}
    next_spk_id = 1
    output_lines = []

    with open(input_stm_path, 'r', encoding='utf-8') as f:
        for line in f:
            line = line.rstrip('\n')
            if not line.strip():
                output_lines.append(line)
                continue

            parts = line.split(maxsplit=6)
            if len(parts) < 7:
                print(f"Skipping malformed line: {line}")
                continue

            filename, channel, speaker, start, end, label, transcript = parts

            if speaker not in speaker_map:
                speaker_map[speaker] = f"spk{next_spk_id}"
                next_spk_id += 1

            new_speaker = speaker_map[speaker]
            new_line = f"{filename} {channel} {new_speaker} {start} {end} {label} {transcript}"
            output_lines.append(new_line)

    with open(output_stm_path, 'w', encoding='utf-8') as f_out:
        f_out.write("\n".join(output_lines) + "\n")

    #write the dictionary tsv file
    with open(dict_tsv_path, 'w', encoding='utf-8') as f_dict:
        f_dict.write("old_speaker_id\tnew_speaker_id\n")
        for old_spk, new_spk in speaker_map.items():
            f_dict.write(f"{old_spk}\t{new_spk}\n")

    print(f"Replaced speakers in {output_stm_path}")
    print(f"Speaker dictionary saved to {dict_tsv_path}")

create_speaker_mapping_and_replace(input_stm, output_stm, dict_tsv)
```
