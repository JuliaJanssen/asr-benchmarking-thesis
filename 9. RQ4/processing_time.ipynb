```python
from pathlib import Path
import os
import pandas as pd
import numpy as np
```

```python
models_root = Path("path/to/your/data") ###
```

```python
matrix = {}

for model_dir in models_root.iterdir():
    if not model_dir.is_dir():
        continue

    tsv_path = model_dir / "tsv"

    for hyp_file in tsv_path.glob("*.tsv"):
        df = pd.read_csv(hyp_file, sep="\t", header=0)
        col = df["RTF"]

        mean_val = col.mean()
        std_val = col.std()

        parts = hyp_file.stem.split("_")  
        y_axis = parts[0]
        x_axis = parts[1]
        entry = f"{mean_val:.2f} ({std_val:.2f})"

        if y_axis not in matrix:
            matrix[y_axis] = {}
        matrix[y_axis][x_axis] = entry

df_matrix = pd.DataFrame(matrix).T.sort_index(axis=0).sort_index(axis=1)
df_numeric = df_matrix.applymap(lambda s: float(s.split(" ")[0]) if pd.notnull(s) else np.nan)
row_mean = df_numeric.mean(axis=1)
row_std = df_numeric.std(axis=1)
df_matrix["avg"] = [f"{m:.2f} ({s:.2f})" for m, s in zip(row_mean, row_std)]
df_matrix = df_matrix.loc[row_mean.sort_values(ascending=False).index]

print(df_matrix)
```
